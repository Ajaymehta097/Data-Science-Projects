```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+mysqlconnector://root:root@localhost/covid_project"
)
```

```python
data= pd.read_csv("E:\data science\Projects\Project2\italy_covid19_sample.csv")

data.to_sql(
    name='covid_data',
    con=engine,
    if_exists='replace',
    index=False
)
```

```python
query = """
select date,new_confirmed_cases from covid_data where new_confirmed_cases > 300 order by date
"""
data_sql=pd.read_sql(query,engine)
data_sql.head()
```
